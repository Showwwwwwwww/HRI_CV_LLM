### Attention rate

**Attention rate** = *Participant count* / *Passerby count*

```python
observations_oc
```

```python
rounding = 2

attention_rate_d = observations_default.Participant.mean()*100
attention_rate_o = observations_oc.Participant.mean()*100
print(f"Participation rates for:\nDefault: {round(attention_rate_d,rounding)}%\nOC-SORT: {round(attention_rate_o,rounding)}%")
```

### Feedback rate

**Feedback rate** = *Feedback count* / *Participant count*

```python
dp = observations_default[(observations_default["Participant"] > 0)]
op = observations_oc[(observations_oc["Participant"] > 0)]
dps = dp[dp.Survey > 0]
ops = op[op.Survey > 0]
feedback_rate_d = len(dps)/len(dp)*100
feedback_rate_o = len(ops)/len(op)*100
print("Feedback rates for:")
print(f"Default: {round(feedback_rate_d,rounding)}%")
print(f"OC-SORT: {round(feedback_rate_o,rounding)}%")
```

### Success rate

**Success rate** = *Successful engagement count* / *Participant count*

```python
success_d = dp.Success.sum()/len(dp)*100
success_o = op.Success.sum()/len(op)*100
print("Success rates for:")
print(f"Default: {round(success_d,rounding)}%")
print(f"OC-SORT: {round(success_o,rounding)}%")
```

## Two Proportions Test for binary data

## Preamble

```python
from statsmodels.stats.proportion import proportions_ztest

# Only analyse data with participant == 1, except for participation rate
# Also encode Emotion column to 1's and 0's
# This has the effect of changing Emotion to Emotion_P
binary_data_names_o = ["Participant", "Success", "Invested", "Photo", "Emotion", "Survey"]
o_oc_p = pd.get_dummies(observations_oc[observations_oc.Participant==1][binary_data_names_o], columns=["Emotion"], drop_first=True)
o_d_p = pd.get_dummies(observations_default[observations_default.Participant==1][binary_data_names_o], columns=["Emotion"], drop_first=True)
```

### Does robot behaviour influence participation rate?

```python
c = np.array([o_oc_p.shape[0], o_d_p.shape[0]])
t = np.array([observations_oc.shape[0], observations_default.shape[0]])
print("Is there a statistical difference in terms of participation rate between OC-SORT and Default behaviour?")
pstat, pp = proportions_ztest(c,t)
print("Final = " + '{0:0.5f}'.format(pp))
print("There is a significant difference" if pp<0.05 else "There is no significant difference")
```

### Is there significance in other binary data?

```python
bd_names = o_d_p.columns.to_list()[1:]
oc_c = o_oc_p.sum().to_numpy()[1:] # Skip participant column
d_c = o_d_p.sum().to_numpy()[1:]
oc_t = np.array([o_oc_p.shape[0] for _ in range(len(bd_names))])
d_t = np.array([o_d_p.shape[0] for _ in range(len(bd_names))])

stacks_c = np.column_stack((oc_c, d_c))
stacks_t = np.column_stack((oc_t, d_t))
```

```python
for i, n in enumerate(bd_names):
  print("Two proportions test for", n)
  v, s = proportions_ztest(stacks_c[i],stacks_t[i])
  print("p = " + '{0:0.5f}'.format(s))
  print("There is a significant difference" if s<0.05 else "There is no significant difference" + "\n")
```

### Tests for non-participants

```python
no_oc_p = pd.get_dummies(observations_oc[observations_oc.Participant==0][binary_data_names_o], columns=["Emotion"], drop_first=True)
no_d_p = pd.get_dummies(observations_default[observations_default.Participant==0][binary_data_names_o], columns=["Emotion"], drop_first=True)
```

```python
noc_c = no_oc_p.sum().to_numpy()[1:] # Skip participant column
nd_c = no_d_p.sum().to_numpy()[1:]
noc_t = np.array([no_oc_p.shape[0] for _ in range(len(bd_names))])
nd_t = np.array([no_d_p.shape[0] for _ in range(len(bd_names))])

nstacks_c = np.column_stack((noc_c, nd_c))
nstacks_t = np.column_stack((noc_t, nd_t))

for i, n in enumerate(bd_names):
  print("Two proportions test for", n)
  v, s = proportions_ztest(nstacks_c[i],nstacks_t[i])
  print("p = " + '{0:0.5f}'.format(s))
  print("There is a significant difference" if s<0.05 else "There is no significant difference" + "\n")
```
